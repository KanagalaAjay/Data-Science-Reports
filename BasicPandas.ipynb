```python
emp = {'id':[12,13,14,15],
      'name':['Allen','Bric','Calvin', 'Don']}

df_emp = pd.DataFrame(emp, columns=['id','name'])
df_emp
```

```python
sal = {'id':[14,15,16,17],
       'salary':['7264','7236','8357','33053']}

df_sal = pd.DataFrame(sal, columns=['id','salary'])
df_sal
```

```python
pd.merge(df_emp, df_sal, on='id') #inner join
```

```python
pd.merge(df_emp, df_sal, on='id', how='outer')
```

```python
pd.merge(df_emp, df_sal, on='id', how='left')
```

```python
pd.merge(df_emp, df_sal, on='id', how='right')
```
